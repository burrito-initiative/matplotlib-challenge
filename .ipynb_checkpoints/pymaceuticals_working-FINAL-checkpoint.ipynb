```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# File to Load (Remember to Change These)
mouse_drug_data_csv = "data/mouse_drug_data.csv"
clinical_trial_data_csv = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_csv)
clinical_trial_data = pd.read_csv(clinical_trial_data_csv)

# Combine the data into a single dataset
clinical_df = pd.merge(mouse_drug_data,clinical_trial_data, on="Mouse ID")

# Display the data table for preview
clinical_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_vol = clinical_df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].agg("mean")

# Convert to DataFrame
mean_tumor_vol = pd.DataFrame(mean_tumor_vol)
mean_tumor_vol = mean_tumor_vol.rename(columns={"Tumor Volume (mm3)":"Tumor Volume (mean)"})

# Preview DataFrame
mean_tumor_vol.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_standard_error = clinical_df.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]

# Convert to DataFrame
tumor_standard_error = pd.DataFrame(tumor_standard_error)
tumor_standard_error = tumor_standard_error.rename(columns={"Tumor Volume (mm3)":"Tumor Volume (SEM)"})

# Preview DataFrame
tumor_standard_error.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_master = mean_tumor_vol.unstack("Drug")
tumor_standard_error = tumor_standard_error.unstack("Drug")

# Preview that Reformatting worked
pivot_master
# tumor_standard_error
```

```python
# Preview dataframe column headers for drug names
drug_names = []
for col in pivot_master.columns:
    drug_names.append(col)
# drug_names
```

```python
# Generate the Plot (with Error Bars)
x_axis_tumor = np.arange(0, 46, 5)
fig, ax = plt.subplots()

# Universal formatting values for Plot Data section
# > > > > > > > > > > > >
markersize = 6
linestyle = "--"
linewidth = 1

# Plot Data
# > > > > > > > > > > > >
# Capomulin
ax.errorbar(x_axis_tumor, 
            pivot_master[('Tumor Volume (mean)', 'Capomulin')],
            tumor_standard_error[('Tumor Volume (SEM)', 'Capomulin')],
            marker="o",markersize=markersize,linestyle=linestyle,linewidth=linewidth,label='Capomulin')
# Infubinol 
ax.errorbar(x_axis_tumor, 
            pivot_master[('Tumor Volume (mean)', 'Infubinol')],
            tumor_standard_error[('Tumor Volume (SEM)', 'Infubinol')],
            marker="s",markersize=markersize,linestyle=linestyle,linewidth=linewidth,label='Infubinol')
# Ketapril
ax.errorbar(x_axis_tumor, pivot_master[('Tumor Volume (mean)', 'Ketapril')],
            tumor_standard_error[('Tumor Volume (SEM)', 'Ketapril')],
            marker="^",markersize=markersize,linestyle=linestyle,linewidth=linewidth,label='Ketapril')
# Placebo
ax.errorbar(x_axis_tumor, 
            pivot_master[('Tumor Volume (mean)', 'Placebo')],
            tumor_standard_error[('Tumor Volume (SEM)', 'Placebo')],
            marker="*",markersize=markersize,linestyle=linestyle,linewidth=linewidth,label='Placebo')

# Diagram formatting
# > > > > > > > > > > > >
plt.grid(axis='y')
plt.title("Tumor Response to Treatment")
plt.xlabel("Days")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(loc="best", fontsize="small", fancybox=True)

# Save the Figure
# > > > > > > > > > > > >
plt.savefig("images/Tumor_Response_to_Treatment.png")

# Show Figure
# > > > > > > > > > > > >
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met_site = clinical_df.groupby(["Drug","Timepoint"])["Metastatic Sites"].agg("mean")

# Convert to DataFrame
mean_met_site = pd.DataFrame(mean_met_site)
mean_met_site = mean_met_site.rename(columns={"Metastatic Sites":"Metastatic Sites (mean)"})

# Preview DataFrame
mean_met_site.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_standard_error = clinical_df.groupby(["Drug", "Timepoint"]).sem()["Metastatic Sites"]

# Convert to DataFrame
met_standard_error = pd.DataFrame(met_standard_error)
met_standard_error = met_standard_error.rename(columns={"Metastatic Sites":"Metastatic Sites (SEM)"})

# Preview DataFrame
met_standard_error.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
met_pivot_master = mean_met_site.unstack("Drug")
met_standard_error  = met_standard_error.unstack("Drug")

# Preview that Reformatting worked
met_pivot_master
met_standard_error
```

```python
# Generate the Plot (with Error Bars)
x_axis_met = np.arange(0, 46, 5)
fig, ax = plt.subplots()

# Universal formatting values for Plot Data section
# > > > > > > > > > > > >
markersize = 6
linestyle = "--"
linewidth = 1

# Plot Data
# > > > > > > > > > > > >
# Capomulin
ax.errorbar(x_axis_met, 
            met_pivot_master[('Metastatic Sites (mean)', 'Capomulin')],
            met_standard_error[('Metastatic Sites (SEM)', 'Capomulin')],
            marker="o",markersize=markersize,linestyle=linestyle,linewidth=linewidth,label='Capomulin')
# Infubinol 
ax.errorbar(x_axis_met, 
            met_pivot_master[('Metastatic Sites (mean)', 'Infubinol')],
            met_standard_error[('Metastatic Sites (SEM)', 'Infubinol')],
            marker="s",markersize=markersize,linestyle=linestyle,linewidth=linewidth,label='Infubinol')
# Ketapril
ax.errorbar(x_axis_met, met_pivot_master[('Metastatic Sites (mean)', 'Ketapril')],
            met_standard_error[('Metastatic Sites (SEM)', 'Ketapril')],
            marker="^",markersize=markersize,linestyle=linestyle,linewidth=linewidth,label='Ketapril')
# Placebo
ax.errorbar(x_axis_met, 
            met_pivot_master[('Metastatic Sites (mean)', 'Placebo')],
            met_standard_error[('Metastatic Sites (SEM)', 'Placebo')],
            marker="*",markersize=markersize,linestyle=linestyle,linewidth=linewidth,label='Placebo')

# Plot formatting
# > > > > > > > > > > > >
plt.grid(axis='y')
plt.title("Metastatic Response to Treatment")
plt.xlabel("Days")
plt.ylabel("Metastatic Sites")
plt.legend(loc="best", fontsize="small", fancybox=True)

# Save the Figure
# > > > > > > > > > > > >
plt.savefig("images/Metastatic_Response_to_Treatment.png")

# Show Figure
# > > > > > > > > > > > >
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count = clinical_df.groupby(["Drug", "Timepoint"]).nunique()["Mouse ID"]

# Convert to DataFrame
mice_count = pd.DataFrame(mice_count)
mice_count = mice_count.rename(columns={"Mouse ID":"Mice Count"})

# Preview DataFrame
mice_count.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_count_pivot_master = mice_count.unstack("Drug")

# Preview the Data Frame
mice_count_pivot_master
```

```python
# Generate the Plot (Accounting for percentages)
x_axis_mice = np.arange(0, 46, 5)

# Calculate percentages
Capomulin_survival_percent = (mice_count_pivot_master.iloc[:, [0]]/mice_count_pivot_master.iloc[:, [0]].iloc[0])*100
Infubinol_survival_percent = (mice_count_pivot_master.iloc[:, [2]]/mice_count_pivot_master.iloc[:, [2]].iloc[0])*100
Ketapril_survival_percent = (mice_count_pivot_master.iloc[:, [3]]/mice_count_pivot_master.iloc[:, [3]].iloc[0])*100
Placebo_survival_percent = (mice_count_pivot_master.iloc[:, [5]]/mice_count_pivot_master.iloc[:, [5]].iloc[0])*100

# Universal formatting values for Plot Data section
# > > > > > > > > > > > >
markersize = 6
linestyle = "--"
linewidth = 1

# Plots
plt.plot(x_axis_mice,Capomulin_survival_percent,
        marker="o",markersize=markersize,linestyle=linestyle,linewidth=linewidth,label='Capomulin')

plt.plot(x_axis_mice,Infubinol_survival_percent,
        marker="s",markersize=markersize,linestyle=linestyle,linewidth=linewidth,label='Infubinol')

plt.plot(x_axis_mice,Ketapril_survival_percent,
        marker="^",markersize=markersize,linestyle=linestyle,linewidth=linewidth,label='Ketapril')

plt.plot(x_axis_mice,Placebo_survival_percent,
        marker="*",markersize=markersize,linestyle=linestyle,linewidth=linewidth,label='Placebo')

# Plot formatting
# > > > > > > > > > > > >
plt.grid(axis='both')
plt.title("Mouse Survival Rates")
plt.xlabel("Days")
plt.ylabel("Survival Rate (%)")
plt.legend(loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig("images/Mouse_Survival_Rates.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Creating a bar graph that compares the total % tumor volume change for each drug across the full 45 days.
# Calculate the percent changes for each drug
percent_changes = round((pivot_master.pct_change(periods=9).dropna())*100, 2)

# Display the data to confirm
percent_changes
```

```python
# Store all Relevant Percent Changes into a Tuple
drug_names = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
percent_change = [-19.48,46.12,57.03,51.3]

x_axis_summary = drug_names
y_axis_summary = percent_change

plt.bar(x_axis_summary, y_axis_summary, align="center")
# Splice the data between passing and failing drugs

threshold = 0
above_threshold = np.maximum(percent_change, threshold)
below_threshold = np.minimum(percent_change, threshold)

plt.bar(x_axis_summary, below_threshold, width=0.8, color="g")
plt.bar(x_axis_summary, above_threshold, width=0.8, color="r")

# Orient widths. Add labels, tick marks, etc. 
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("Tumor Volume % Change")
plt.grid(axis='y')

tick_locations = [value for value in x_axis_summary]
plt.xticks(tick_locations, drug_names)

# Save the Figure
plt.savefig("images/Tumor_Change_After_45_Days.png") 

# Show the Figure
fig.show()
```
